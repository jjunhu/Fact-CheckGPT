```python
import matplotlib.pyplot as plt
from collections import defaultdict

# Read the training loss log data from the file
training_loss_filename = 'training_loss_log_batch_size_32.txt'
validation_loss_filename = 'validation_loss_log_batch_size_32.txt'

def read_loss_file(filename):
    epoch_losses = defaultdict(list)
    with open(filename, 'r') as f:
        for line in f:
            parts = line.strip().split(',')
            epoch = int(parts[0].split(':')[1].strip())
            loss = float(parts[2].split(':')[1].strip())
            epoch_losses[epoch].append(loss)
    return epoch_losses

def compute_average_losses(epoch_losses):
    epochs = sorted(epoch_losses.keys())
    avg_losses = [sum(epoch_losses[epoch])/len(epoch_losses[epoch]) for epoch in epochs]
    return epochs, avg_losses

# Read and compute average losses for training and validation logs
training_epoch_losses = read_loss_file(training_loss_filename)
validation_epoch_losses = read_loss_file(validation_loss_filename)

training_epochs, avg_training_losses = compute_average_losses(training_epoch_losses)
validation_epochs, avg_validation_losses = compute_average_losses(validation_epoch_losses)

# Plotting the average training and validation loss per epoch
plt.figure(figsize=(10, 6))
plt.plot(training_epochs, avg_training_losses, label='Average Training Loss')
plt.plot(validation_epochs, avg_validation_losses, label='Average Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Average Training and Validation Loss Batch Size = 32 per Epoch')
plt.legend()
plt.grid(True)
plt.show()

```
